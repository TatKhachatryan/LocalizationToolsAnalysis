```python
import pandas as pd
import json

# Provide the file path to your JSON file
file_path = 'G2_reviews_transifex.json'
company = 'Transifex'
# Open and read the JSON file
with open(file_path, 'r') as file:
    data = json.load(file)

print(data.keys())
```

```python
# Create a DataFrame from the 'reviews' list
df = pd.DataFrame(data['reviews'])
df['product_name'] = data['product_name']
df
```

```python
# Remove the unwanted sentence
df['review'] = df['review'].str.replace(
    "Review collected by and hosted on G2.com.", "", regex=False).str.strip()
```

```python
# Split the "reviews" column into "Pros" and "Cons"
df[['Pros', 'Cons']] = df['review'].str.extract(
    rf"What do you like best about {company}\?\s*(.*?)\s*What do you dislike about {company}\?\s*(.*)")
```

```python
df['summary'].value_counts()
```

```python
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df.drop(['summary', 'review'], axis = 1, inplace=True) 
```

```python
df.to_csv('G2_reviews_Transifex.csv', index=False)
```
